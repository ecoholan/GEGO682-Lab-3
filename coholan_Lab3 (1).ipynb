```python
import rasterio
from rasterio import plot
from rasterio.merge import merge
import matplotlib.pyplot as plt
```

```python
satdat1 = rasterio.open("image1.tif")
satdat2 = rasterio.open("image2.tif")
```

```python
plot.show(satdat1, cmap='terrain')
```

```python
plot.show(satdat2, cmap='terrain')
```

WGS 84 / UTM zone 18N

```python
fig, (axr, axg, axb) = plt.subplots(1,3, figsize=(20,5))
plot.show((satdat1, 3), ax=axr, cmap='Reds', title='Red Channel')
plot.show((satdat1, 2), ax=axg, cmap='Greens', title='Green Channel')
plot.show((satdat1, 1), ax=axb, cmap='Blues', title='Blue Channel')
```

```python
files_to_mosaic = [satdat1, satdat2]
```

```python
mosaic, out_trans = merge(files_to_mosaic)
```

```python
plot.show(mosaic, cmap = 'terrain')
```
